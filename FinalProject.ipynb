Final Project - Interactive HRD of Matching RSG Candidates
- 
- For my final project, I created a module that would help me solve and organize a problem in my research, on locating red supergiants (RSGs) in NGC 6822. My overall motivation was to gain some insight into determining specific temperature and luminosity constraints for my proposed RSG sample, and my specific motivation was to simplify and organize this process in order to be able to easily utilize it in the future. Something that would help with this problem is being able to compare my proposed sample against other samples of confirmed RSGs in NGC 6822, by finding matches between datasets and plotting the matches on an HR-diagram. To solve and organize this problem, I created a module with a function that takes an input of two datasets, and a desired angle of separation, to find matches between them. Then, I added two functions that are able to plot a dataset and its matched stars on an HR-diagram, with an interactive widget to serve as a visual tool for presenting the dataset and the matched results. 
- Solution: creating and importing a module with three functions: matching stars between two datasets, HRD of the results, and an interactive widget for the HRD

Part 1.) Matching Stars Function
- In the interactive_match module, I created the match_stars function, which allows me to input the filename of my dataset and the filename of a comparison datset, and an angle of separation, to get an organized table of matching stars as an output. The function singles out and converts columns of RA & DEC to skycoord degrees. Then, it identifies the id of the matched stars and pulls them from my dataset, and returns a table of the identified stars with new columns of angular separation and distance.

```python
import interactive_match
```

```python
interactive_match.match_stars('RSG1.csv','RSGpaper1.csv', 2.0)
```

Part 2.) Interactive HR-Diagram Widget
-
- In the interactive_match module, I created the hrd and interactive_hrd function. The hrd function calculates the temperatures and luminosities of my dataset (in pink), and of the matched stars (in blue), then plots an HR-diagram. The interactive_hrd function creates sliders that can toggle the point sizes of each dataset, in order for the plot to highlight or put emphasis on one dataset over the other for future visual presentation purposes. 

```python
matched = interactive_match.match_stars('RSG1.csv','RSGpaper1.csv', 2.0)
```

```python
interactive_match.interactive_hrd('RSG1.csv', matched, 6, 4)
```
